```python
pi = 4.0 * np.mean(np.sqrt(1 - x**2))
print(pi)
```

```python
import matplotlib.pyplot as plt
np.random.seed(1)
N = [10,20, 40, 50, 100, 200, 500, 1000]
pis = []
pis_var = []

M = 20
for n in N:
    pin = []
    for m in range(M):
        np.random.seed(n+m)
        x = np.random.rand(n)
        pi = 4.0 * np.mean(np.sqrt(1 - x**2))
        pin.append(pi)
    pis.append(np.array(pin).mean())
    pis_var.append(np.array(pin).var())
    # np.random.seed(n+m)
    # x = np.random.rand(n)
    # pi = 4.0 * np.mean(np.sqrt(1 - x**2))
    # pis.append(pi)

fig, ax = plt.subplots()
ax.plot(N, pis, 'o-')
ax.fill_between(N, np.array(pis) - np.sqrt(pis_var), np.array(pis) + np.sqrt(pis_var), alpha=0.2)
ax.set_xlabel('Number of samples')
ax.set_ylabel('pi approx')
ax.axhline(np.pi, color='red', ls='--')
ax.grid(True)
```
