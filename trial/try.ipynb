```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

np.random.seed(42)
# Generate 100 samples from uniform distribution
samples = np.random.uniform(size=1000)
sns.kdeplot(samples, label='KDE of Samples', fill=True)

# Plot the samples
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

np.random.seed(41)
# Generate 100 samples from uniform distribution

def trf1(x):
    return x**4

samples = np.random.uniform(size=2000)
# sns.kdeplot(trf1(samples), label='KDE of Samples', fill=True)
plt.hist(trf1(samples), bins=100, density=True, label='KDE of Samples', fill=True)

# Plot the samples
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
